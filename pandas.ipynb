```python
n_rows = 10
n_cols = 2
cols = ["omset", "operasional"]
dfBaru = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
dfBaru
```

```python
dfBaru["omset"] = dfBaru["omset"] * 100_00
dfBaru["operasional"] = dfBaru["operasional"] * 10_000
dfBaru
```

```python
dfBaru.index = pd.util.testing.makeDateIndex(10, freq="D")
dfBaru = dfBaru.reset_index()
dfBaru = dfBaru.rename(columns={"index":"tanggal"})
dfBaru
```

```python
formatku = {"tanggal" :"{:%d/%m/%y}",
            "omset": "Rp.{:.2f}",
            "operasional":"Rp.{:.2f}"}
laporan = dfBaru.style.format(formatku)
laporan.highlight_min("omset", color="pink")
laporan.highlight_max("omset", color="gray")
laporan.highlight_min("operasional", color="pink")
laporan.highlight_max("operasional", color="gray")
laporan.set_caption("WOW")
```

```python
dfBaru["Hasil"] = dfBaru["omset"].cummax()
dfBaru
```
