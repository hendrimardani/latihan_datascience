```python
import pandas as pd
import tensorflow as tf
import numpy as np
import sklearn.preprocessing as pre
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
data = pd.read_csv('citrus.csv')
data
```

```python
name = pre.LabelEncoder()
name.fit(data["name"])
data["name"] = name.transform(data["name"])

# Ubah type data
data["name"] = data["name"].astype("float32")
data
```

# UBAH MENJADI ARRAY

```python
dataset = data.values
```

# MELAKUKAN NORMALISASI DATA

```python
X = dataset[:, 1:]
y = dataset[:, 0]

skala = pre.MinMaxScaler()
X = skala.fit_transform(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# MEMBUAT 3 LAYERS

```python
# Pada input shape diberikan 5 karena kita memiliki 5 Features
# Pada activation function pilih sigmoid karena probabilitas 0 dan 1 Cocok untuk klasifikasi biner
model = Sequential([    
                    Dense(32, activation='relu', input_shape=(5,)),    
                    Dense(32, activation='relu'),    
                    Dense(1, activation='sigmoid')
])
```

# MEMBUAT COMPILE

```python
# Pada optimizer pilih "sgd", karena cocok untuk dataset berukuran kecil
model.compile(optimizer='sgd', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

# MENGEVALUASI MODEL

```python
# Elemen pertama  loss, Elemen kedua akurasi
model.evaluate(X_test, y_test)
```
