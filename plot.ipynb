```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("titanic.csv")
data
```

```python
data.describe()
```

```python
data["age"].hist()
```

```python
sns.jointplot(x="age", y="fare", color="red", hue="sex", data=data);
```

```python
from scipy.stats import zscore
import numpy as np
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
nilaiZ = zscore(data.fare)
nilaiZ
```

```python
nilaiBaru = np.abs(nilaiZ)
nilaiBaru
```

```python
filterData = nilaiBaru < 3
dataBaru = data[filterData]
dataBaru
```

```python
sns.jointplot(x="age", y="fare", hue="who", height=8, data=dataBaru);
plt.savefig("a.png")
```

```python
sns.kdeplot(x="age", y="fare", shade=True, cbar=True, data=data);
```

```python
sns.boxplot(x="deck", y="fare", hue="who", order=["A", "B", "C", "D", "E", "F", "G"], data=data);
```

```python
sns.barplot(x="who", y="age", data=data);
```
